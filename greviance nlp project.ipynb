```python
import string
string.punctuation 
```

```python
def text_process(mess):            ### creating a function

    """                                                        ## a docstring
    1. remove the punctuation
    2. remove the stopwords
    3. return the list of clean textwords
    """
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = "".join(nopunc)
    return [ word for word in nopunc.split() if word not in l1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bow_trasformer = CountVectorizer(analyzer= text_process).fit(gre_df["X"])

bow_trasformer.vocabulary_
```

```python
len(bow_trasformer.vocabulary_)
```

```python
gre_df_bow = bow_trasformer.transform(gre_df.X)
```

```python
type(gre_df_bow)
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(gre_df_bow,gre_df.Y,random_state=123,test_size=.2)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score,f1_score
```

```python
from sklearn.linear_model import LogisticRegression

logreg  = LogisticRegression()
logreg.fit(train_x,train_y)
pred    = logreg.predict(test_x)

from sklearn.metrics import confusion_matrix,classification_report

tab = confusion_matrix(test_y,pred)
print(tab)
print("-----------------------------------")
print(classification_report(test_y,pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

nb=MultinomialNB()
nb.fit(train_x,train_y)
pred_nb= nb.predict(test_x)

from sklearn.metrics import confusion_matrix, classification_report

tab_nb = confusion_matrix(test_y,pred_nb)
print(tab_nb)
print("--------------------------------------")
print(classification_report(test_y,pred_nb))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(train_x , train_y)
pred_dt = dt.predict(test_x)

from sklearn.metrics import confusion_matrix, classification_report

tab_dt = confusion_matrix(test_y,pred_dt)
print(tab_dt)
print("--------------------------------------")
print(classification_report(test_y,pred_dt))
```
